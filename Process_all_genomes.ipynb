```python
import os
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
def reverse_complement(seq):
    return str(Seq(seq).reverse_complement())

#Directory qui contient les génomes des bactéries
base_directory = "/home/seddik/Documents/Synesth/iter4/Bacteria"
#On va utiliser ces noms pour join et trouver le actual bon path
bacteria_names = ["Thioalkalivibrio", "ClostridiumKluyveri", "Halodurans", "Staph_Epidermidis", "Rhodococcus", "Cyanothece", "Synechocystis", "Thermotoga", "Candidatus", "Geobacter", "Shewanella", "Vibrio", "Yersinia", "Ecoli"]


#Permet d'extraire le bon gène sous le bon format pour une certaine séquence. 
def extract_gene_sequence(record, start, length, orientation):
    sequence = record.seq[(start-1):(start-1+length)]
    if orientation == 2:
        sequence = reverse_complement(sequence)
    return str(sequence)
    
gene_sequences = {}

#On itère à travers les directories 
for name in bacteria_names:
        
    bacteria_dir = os.path.join(base_directory, name)

    #on a un seul file par dossier. Donc cette liste contient un seul élément.
    fasta_files = [f for f in os.listdir(bacteria_dir)]
    
    #error catching
    if not fasta_files:
        print(f"No FASTA file found in directory: {bacteria_dir}")
        continue
    
    #On obtient le actual absolute path de chaque GCA file (contenant le complete génome)
    fasta_file = os.path.join(bacteria_dir, fasta_files[0])

    # Initialise le dictionnaire pour la bactérie considérée

    for synteny in bacteria_syntenies[name]:

        gene_sequences[synteny] = {}

    #Nous permet de connaitre l'indice du record qu'il faut checker dans le génome.
    #Si le bacteria name n'est pas dans le dictionnaire des plasmides, alors on sait
    #que le record voulu est placé dans le chromosome et donc la valeur de 0 par défaut
    #sera donnée à l'indice. 
        record_index = plasmides.get(synteny, 0)

        for i, record in enumerate(SeqIO.parse(fasta_file, "fasta")):
            if i == record_index:
            # on a le bon indice
                for cas_gene, (start, length, orientation) in extraction_dict.get(synteny, {}).items():
                    gene_sequence = extract_gene_sequence(record, start, length, orientation)
                    gene_sequences[synteny][cas_gene] = gene_sequence
                break  # Exit the loop once the desired record is processed
```

```python
from Bio.SeqRecord import SeqRecord

output_directory = "/home/seddik/Documents/Synesth/iter6"

for bacterium, genes in gene_sequences.items():
    for cas_gene, sequence in genes.items():

        if not sequence.strip():
            continue

        cleaned_sequence = ''.join(sequence.split())
                                   
        fasta_file = os.path.join(output_directory, f"all_{cas_gene}.fna")
        seq_record = SeqRecord(Seq(sequence), id=bacterium, description="")
        
        with open(fasta_file, "a") as output_handle:
            SeqIO.write(seq_record, output_handle, "fasta")

```
